### Notation

$\mathbf{X}$ is the $M \times T$ data matrix, containing $M$ measurements at $T$ time steps.

$\mathbf{S}$ is the $S \times T$ source matrix, containing $S$ source signal values at $T$ time steps. We will assume $S = M$.

$\mathbf{A}$ is the mixing matrix. We have $\mathbf{X} = \mathbf{A S}$.

$\mathbf{W}$ is the matrix we aim to learn. It is the inverse of $\mathbf{A}$, up to indeterminacies (scaling and permutation of sources).

$\phi$ is an elementwise non-linearity or activation function, typically applied to elements of $\mathbf{W X}$.

In the following assignments, you can make use of the signal generators listed below.



```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
# Signal generators
def sawtooth(x, period=0.2, amp=1.0, phase=0.):
    return (((x / period - phase - 0.5) % 1) - 0.5) * 2 * amp

def sine_wave(x, period=0.2, amp=1.0, phase=0.):
    return np.sin((x / period - phase) * 2 * np.pi) * amp

def square_wave(x, period=0.2, amp=1.0, phase=0.):
    return ((np.floor(2 * x / period - 2 * phase - 1) % 2 == 0).astype(float) - 0.5) * 2 * amp

def triangle_wave(x, period=0.2, amp=1.0, phase=0.):
    return (sawtooth(x, period, 1., phase) * square_wave(x, period, 1., phase) + 0.5) * 2 * amp

def random_nonsingular_matrix(d=2):
    """
    Generates a random nonsingular (invertible) matrix if shape d*d
    """
    epsilon = 0.1
    A = np.random.rand(d, d)
    while abs(np.linalg.det(A)) < epsilon:
        A = np.random.rand(d, d)
    return A

def plot_signals(X,label=""):
    """
    Plot the signals contained in the rows of X.
    """
    fig = figure()
    for i in range(X.shape[0]):
        ax = plt.subplot(X.shape[0], 1, i + 1)
        plot(X[i, :])
        ax.set_xticks([])
        ax.set_yticks([])
    fig.suptitle(label, fontsize=14, fontweight='bold')
```


### 1.3 Implicit priors (20 points)
As explained in MacKay's book, an activation function $\phi$ used in the ICA learning algorithm corresponds to a prior distribution over sources. Specifically, $\phi(a) = \frac{d}{da} \ln p(a)$. For each of the following activation functions, derive the source distribution they correspond to.
$$\phi_0(a) = -\tanh(a)$$
$$\phi_1(a) = -a + \tanh(a)$$
$$\phi_2(a) = -a^3$$
$$\phi_3(a) = -\frac{6a}{a^2 + 5}$$

The normalizing constant is not required, so an answer of the form $p(a) \propto \verb+[answer]+$ is ok.

Plot the activation functions and the corresponding prior distributions. Compare the shape of the priors to the histogram you plotted in the last question.


#### ANSWER:
0. **$\phi_0(a) = -\tanh(a)$**
$$
-\tanh(a) = \frac{d}{da} \ln p(a) \\
\int -\tanh(a)da \propto -\ln(\cosh(a)) \Rightarrow \\
p(a) \propto  \cosh(a)^{-1}
$$

1. **$\phi_1(a) = -a + \tanh(a)$**
$$
-a + \tanh(a) = \frac{d}{da} \ln p(a) \\
\int -a + \tanh(a) da \propto \ln\left(\cosh(a) \exp \left\{-\frac{a^2}{2} \right\} \right)\Rightarrow \\
p(a) \propto  \cosh(a) \exp \left\{-\frac{a^2}{2} \right\}
$$

2. **$\phi_2(a) = -a^3$**
$$
-a^3 = \frac{d}{da} \ln p(a) \\
\int -a^3 da \propto \frac{-a^4}{4} \Rightarrow \\
p(a) \propto \exp \left\{\frac{-a^4}{4} \right\}
$$

3. **$\phi_3(a) =-\frac{6a}{a^2 + 5}$**
$$
-\frac{6a}{a^2 + 5} = \frac{d}{da} \ln p(a) \\
\int -\frac{6a}{a^2 + 5}  da \propto \log((a^2+5)^{-3}) \Rightarrow \\
p(a) \propto (a^2+5)^{-3}
$$

```python

rcParams['figure.figsize'] = 5, 5
phi_0 = lambda a: - np.tanh(a)
phi_1 = lambda a: -a + np.tanh(a)
phi_2 = lambda a: -a**3
phi_3 = lambda a: -(6*a)/(a**2 + 5)
phis = [phi_0, phi_1, phi_2, phi_3]
for i,phi in enumerate(phis):
    X = np.linspace(-1,1,50)
    Y = phi(X)
    plot(X,Y,label="$\phi_%d$"%i)
    plt.legend()
plt.show()

p_0 = lambda a: np.cosh(a)**(-1)
p_1 = lambda a: np.cosh(a)*np.exp(-(a**2)/2.)
p_2 = lambda a: np.exp(-(a**4)/4.)
p_3 = lambda a: (a**2+5)**(-3)

rcParams['figure.figsize'] = 15, 5
ps = [p_0, p_1, p_2, p_3]
fig,subplt = plt.subplots(1,4)
for i,p in enumerate(ps):
    X = np.linspace(-10,10,50)
    Y = p(X)
    subplt[i].plot(X,Y,label="$p_%d$"%i)
    subplt[i].legend()
fig.show()
```

### 1.4 Whitening (15 points)
Some ICA algorithms can only learn from whitened data. Write a method `whiten(X)` that takes a $M \times T$ data matrix $\mathbf{X}$ (where $M$ is the dimensionality and $T$ the number of examples) and returns a whitened matrix. If you forgot what whitening is or how to compute it, various good sources are available online, such as http://courses.media.mit.edu/2010fall/mas622j/whiten.pdf



```python
# Whitening
def whiten(X):
    n_signals = X.shape[0]
    # Make covariance matrix:
    cov = np.cov(X)
    
    # Do eigenvalue decomposition
    eig_values, eigen_vectors = linalg.eigh(cov)
    
    eig_sqrt = np.diag(1 / np.sqrt(eig_values))
    
    white_matrix = np.dot(np.dot(eigen_vectors,eig_sqrt),eigen_vectors.T)
    
    return np.dot(white_matrix,X)

whitened_x = whiten(X)
```

### 1.7 Independent Component Analysis (25 points)
Implement the covariant ICA algorithm as described in MacKay. Write a function `ICA(X, activation_function, learning_rate)`, that returns the demixing matrix $\mathbf{W}$. The input `activation_function` should accept a function such as `lambda a: -tanh(a)`. Update the gradient in batch mode, averaging the gradients over the whole dataset for each update. Try to make it efficient, i.e. use matrix operations instead of loops where possible (loops are slow in interpreted languages such as python and matlab, whereas matrix operations are internally computed using fast C code).



```python
import numpy as np
import math
phi_0 = lambda a: - np.tanh(a)
phi_1 = lambda a: -a + np.tanh(a)
phi_2 = lambda a: -a**3
phi_3 = lambda a: -(6*a)/(a**2 + 5)
phis = [phi_0, phi_1, phi_2, phi_3]
```

```python
def ica(x, activation_function, learning_rate, epsilon, runs):
    num_signals = X.shape[0]
    # init W
    w = random_nonsingular_matrix(d=num_signals)
    i = 0

    while i < runs : 
        a = np.dot(w, x)
        z = activation_function(a)
        x_new = np.dot(w.T, a)
        delta = w + np.dot(z, x_new.T)*(1/float(x.shape[1]))
        
        w = w + learning_rate*delta
        if np.max(np.abs(delta)) < epsilon:
            break
            
        i += 1
    return w
```

### 1.10 Excess Kurtosis (20 points)
The (excess) kurtosis is a measure of 'peakedness' of a distribution. It is defined as
$$
\verb+Kurt+[X] = \frac{\mu_4}{\sigma^4} - 3 = \frac{\operatorname{E}[(X-{\mu})^4]}{(\operatorname{E}[(X-{\mu})^2])^2} - 3
$$
Here, $\mu_4$ is known as the fourth moment about the mean, and $\sigma$ is the standard deviation.
The '-3' term is introduced so that a Gaussian random variable has 0 excess kurtosis.
We will now try to understand the performance of the various activation functions by considering the kurtosis of the corresponding priors, and comparing those to the empirical kurtosis of our data.

First, compute analytically the kurtosis of the four priors that you derived from the activation functions before. To do this, you will need the normalizing constant of the distribution, which you can either obtain analytically (good practice!), using computer algebra software (e.g. Sage) or by numerical integration (see scipy.integrate).
